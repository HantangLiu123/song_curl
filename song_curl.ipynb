```python
#attempt to extract info from a song
SONG_URL = f"https://music.163.com{song_url_list[0]}"
```

```python
#from claude ai
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--enable-javascript')

driver = webdriver.Chrome(options=chrome_options)
driver.get(SONG_URL)
wait = WebDriverWait(driver, 15)

#change to iframe 
iframe = wait.until(EC.presence_of_element_located((By.TAG_NAME, "iframe")))
driver.switch_to.frame(iframe)

#get the image
image = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'j-img')))
img_url = image.get_attribute('src')
IMG_PATH = 'C:/Chris Liu/清华/25暑期/python-course/song_curl/saved_info/song_info_curl_exp/img.jpg'
response = requests.get(img_url, stream=True)
with open(IMG_PATH, "wb") as f:
    for chunck in response.iter_content(chunk_size=8192):
        f.write(chunck)

#click the "展开" button for lyrics
more = wait.until(EC.element_to_be_clickable((By.XPATH, "//*[text()='展开']")))
driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", more)
time.sleep(0.5)
more.click()

#get the lyrcis
element = driver.find_element(By.ID, "lyric-content")
lyrics = element.text
driver.quit()

lyrics
```
